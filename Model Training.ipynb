```python
from ultralytics import YOLO

model = YOLO("yolo11n.yaml")  # побудова моделі з нуля

results = model.train(
    data="data.yaml",  # шлях до конфігураційного файлу з описом вибірки
    imgsz=640,  # розмір вхідних зображень
    patience=20,  # кількість епох для ранньої зупинки
    cos_lr=True,  # використання косинусного затухання швидкості навчання
    augment=True,  # використання аугментацій
    device=0,  # використання GPU з індексом 0
    project="vehicle_detection",  # назва проекту
    name="train_nano",  # назва експерименту
)
```
